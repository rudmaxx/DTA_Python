1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)

## Розділ 6 — Генерація випадкових чисел

> **Що повторюємо:** `np.random.seed()`, `np.random.randint()`,

### ✏️ Завдання 6.1

Змоделюй результати маркетингового опитування 50 клієнтів.

1. Встанови `np.random.seed(99)`
2. Згенеруй масив `ages` — вік 50 клієнтів: цілі числа від 18 до 65
3. Згенеруй масив `scores` — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
4. Обріж `scores` так щоб значення були від 1 до 10 — `np.clip(arr, 1, 10)`
5. Виведи:
   - середній вік клієнтів
   - середню оцінку задоволеності
   - кількість клієнтів з оцінкою >= 8 («задоволені»)
   - відсоток задоволених

````python
# Твій код:
np.random.seed(99)
ages = np.random.randint(18, 66, size=50)
scores = (np.random.normal(loc=7, scale=1.5, size=50)).round(1)
print("2. Згенеруй масив ages")
print(ages,"\n")
print("3. Згенеруй масив scores")
print(scores,"\n")
scores = np.clip(scores, 1, 10)
print("4. Обріж scores так щоб значення були від 1 до 10")
print(scores,"\n")
avg = ages.mean()
print("5.1. Середній вік клієнтів: ",avg,"років")
scr_avg = scores.mean().round(2)
print("5.2. Середня оцінка задоволеності: ",scr_avg)
zad = scores[scores >= 8]
print("5.3. Кількість клієнтів з оцінкою >= 8: ",len(zad))
print("5.4. Відсоток задоволених: ",(zad.size) / 50 * 100, "%")
````

## 🏆 Фінальний проєкт — Аналіз даних магазину

> **Використовуємо все з тьюторіалу разом**

Ти аналітик даних невеликого магазину електроніки.
Тобі передали масив продажів за 8 тижнів (56 днів).

Виконай повний аналіз та дай відповіді на запитання керівника.

````python
# Дані — не змінюй цю клітинку
np.random.seed(2024)
base = np.random.randint(200, 800, 56)

# Вихідні продаються краще
weekend_boost = np.array([1 if i % 7 in [5, 6] else 0 for i in range(56)])
sales_56 = (base + weekend_boost * np.random.randint(100, 300, 56)).astype(int)

print('Дані отримано. Масив sales_56 готовий до аналізу.')
print(f'Елементів: {sales_56.size}, форма: {sales_56.shape}')
````

### Запитання 1 — Загальна статистика (Розділ 1, 3)
Виведи:
- Загальна виручка за 56 днів
- Середньоденна виручка (округли до 2 знаків)
- Мінімальний та максимальний денний продаж
- Стандартне відхилення (наскільки продажі «скачуть»)

````python
# Твій код:
total = sales_56.sum()
print('Загальна виручка за 56 днів: ',total,'\n')
print('Середньоденна виручка: ',sales_56.mean().round(2),'\n')
print('Мінімальний денний продаж:',sales_56.min(), 'та максимальний денний продаж: ',sales_56.max(),'\n')
print('Стандартне відхилення:', sales_56.std().round(2))
````

### Запитання 2 — Тижневий аналіз (Розділ 5)
1. Перетвори `sales_56` у матрицю **8×7** (8 тижнів по 7 днів)
2. Знайди виручку кожного тижня (сума по рядках)
3. Знайди найкращий тиждень і виведи його номер (рахуючи з 1) та суму
4. Знайди середньоденну виручку по кожному дню тижня (середнє по стовпцях)
5. Виведи який день тижня в середньому найприбутковіший (0=Пн, 6=Нд)

````python
# Твій код:
sales = sales_56.reshape(8,7)
print('1. Матриця 8x7: ')
print(sales,'\n')

total_week = sales.sum(axis=1)
print('2. Виручка по тижням: ')
print(total_week,'\n')

best_week = total_week.max()
best_week_index = total_week.argmax()
print(f'3. Найкращий тиждень: {best_week_index + 1} і його сумма: {best_week} \n ')


tdm = sales.mean(axis=0).round(2)
print('4. Середньоденна виручка по кожному дню тижня: ')
print(tdm,'\n')

best_day = tdm.argmax()
print(f'5. Найприбутковіший день тижня: {best_day}, середня виручка: {tdm[best_day]:.2f}')
````

````python
# Твій код:
weak_days = sales[sales < 400]
print(f'1. Кількість «слабких» днів — продажі нижче 400 грн: {len(weak_days)}')

good_days = sales[sales > 700]
print(f'2. Кількість «сильних» днів — продажі вище 700 грн: {len(good_days)}')

avg_weak = weak_days.mean().round(2)
avg_good = good_days.mean().round(2)
print(f'3. Середня виручка у «слабкі» дні: {avg_weak} грн\n   Середня виручка у «сильні» дні: {avg_good} грн' )

labels = np.where(sales_56 < 400,'Слабкий', np.where(
        sales_56 > 700,
        'Сильний',
        'Нормальний')
)
print("5. Перші 14 днів:")
for i in range(14):
    print(f"День {i+1}: {sales_56[i]} - {labels[i]}")
````
