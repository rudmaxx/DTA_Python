Типи результатів функцій

````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

Функція щось виконала, проте цей результ "підставити" ніде не можна

NoneType - нам показали результат (або ні), у будь-якому випадку, працювати з результатом повернення функції ми не можемо

````python
type(None)
````

````python
# Повертає NoneType, без зміни даних, виводить результат
def info(d):
  print(d)

# info(my_dict)
type(info(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, виводить результат
def add_el_view(d):
  d[6] = "six"
  print(d)

# info(my_dict)
type(add_el_view(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, НЕ виводить результат
def add_el(d):
  d[6] = "six"

print(type(add_el(my_dict)))
print(f"my_dict: {my_dict}")
````

````python
print(type(add_el(my_dict)))
display(type(add_el(my_dict)))
````

Функція повертає результат для обробки

````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

````python
def dict_value(d, key):
  return d[key]  # my_dict[3]

  # for k, value in d.items():
  #   if value == key:
  #     return k

print(dict_value(my_dict, 3))
print(type(dict_value(my_dict, 3)))
dict_value(my_dict, 3) + "333"
````

Аргументи

````python
def dict_value(d: dict, key: int) -> str:
  return d[key]  # my_dict[3]

print(dict_value(my_dict, 3))
print(type(dict_value(my_dict, 3)))
dict_value(my_dict, 3) + "333"
````

````python
def func_3(a=5, b=7):
  print(a + b)

func_3()  # 5 + 7
func_3(10)  # 10 + 7 (a=10)
func_3(10, 20)  # 10 + 20 (a=10, b=20)
func_3(b=10)  # 5 + 10 (b=10)
````

````python
def func_4(a, b=7):
  print(a + b)

# func_4()  # TypeError: func_4() missing 1 required positional argument: 'a'
func_4(10)  # 10 + 7 (a=10)
func_4(10, 20)  # 10 + 20 (a=10, b=20)
````
